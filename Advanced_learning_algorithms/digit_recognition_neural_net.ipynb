```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
```

```python
# NN model defined.
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(units = 25, activation = "relu"),
    Dense(units = 15, activation = "relu"),
    Dense(units = 10, activation = "softmax"),
])
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
# Binary Crossentropy loss function for binary classification
model.compile(loss=SparseCategoricalCrossentropy())
```

```python
#loading mnist data set from available datasets.
from tensorflow.keras.datasets import mnist
(X_train,y_train), (X_test, y_test)= mnist.load_data()
```

```python
print(X_train.shape) 
print(y_train.shape) 
print(X_test.shape) 
print(y_test.shape) 
```

```python
model.fit(X_train,y_train,epochs=15)
```

```python
accuracy = model.evaluate(X_test,y_test)
prediction = model.predict(X_test)
print(accuracy)
print(prediction[:,:].shape , prediction.size)
```

```python

(m , n) = prediction[:,:].shape

import numpy as np
for i in range(m):
    print("largest probability value for ", i , "test data = " ,  np.max(prediction[i]) , "class = ", np.argmax(prediction[i]))
```
